```python
import numpy as np
import pandas as pd
from typing import Tuple
```

```python
def sqrt_uneven_dist(center: Tuple[float, float], radius: float) -> list[int]:
    angle = np.random.rand() * 2 * np.pi
    r = radius * np.random.rand()
    return [r * np.cos(angle) + center[0], r * np.sin(angle) + center[1]]
```

```python
def sqrt_uneven_dist_ring(center: Tuple[float, float], radius: Tuple[float, float]) -> list[int]:
    angle = np.random.rand() * 2 * np.pi
    r = radius[0] + (radius[1] - radius[0]) * np.sqrt(np.random.rand())
    return [r * np.cos(angle) + center[0], r * np.sin(angle) + center[1]]
```

```python
x1 = np.linspace(-7, 7, 160)
x2 = np.linspace(-7, 7, 240)
x1 = np.concatenate([x1, x2, x1, x2])
y1 = np.random.normal(size=400, scale=.7) + 4
y2 = np.random.normal(size=400, scale=.7) - 4

df = pd.DataFrame({
    'x': x1,
    'y': np.concatenate([y1, y2])
})
```

```python
import plotly.graph_objects as go

plt = go.Figure()
plt.add_trace(
    go.Scatter(
        x=df['x'],
        y=df['y'],
        mode='markers'
    )
)
plt.show()
```

```python
df.to_csv('test3.csv')
```

```python
for i in range(3, 4):
    df = pd.read_csv(f'test{i}.csv')
    fig = go.Figure()
    mn = min(min(df['y']), min(df['x']))
    mx = max(max(df['y']), max(df['x']))
    fig.update_yaxes(range=[mn - .2, mx + .2])
    fig.update_xaxes(range=[mn - .2, mx + .2])
    fig.add_trace(
        go.Scatter(
            mode='markers',
            x=df['x'],
            y=df['y']
        )
    )
    fig.update_layout(
    autosize=False,
    width=600,
    height=600
    )
    fig.write_image(f'test{i}.png', width=1024, height=1024, scale=4)
```
